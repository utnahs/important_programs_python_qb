```python
# WAP to find the average of all numbers in a list
def avg(input):
    len=0
    sum=0
    for i in input:
        len+=1
        sum+=i
    return round((sum/len),2)
avg([1,2,3,4])   
```

```python
#WAP to find the sample standard deviation of given numbers in list
def stddev(listt):
    count=0
    mean=avg(listt)
    sq_diff_sum = 0
    for i in listt:
        count+=1
        sq_diff_sum+=(i-mean)**2
    s=(sq_diff_sum/(count-1))**0.5
    return s
stddev([1,2,3,4,5])
```

```python
#WAP to check whether two strings are anagrams

def is_prime(num):
    c=0
    for i in range(1,num+1):
        if num%i==0:
            c+=1
    if c==2:
        return True
    else:
        return False
    
def prime_list():
    numbers=2
    c=1
    list_of_prime=[]
    while c<=26:
        if is_prime(numbers)==True:
            list_of_prime.append(numbers)
            c+=1
        numbers+=1
    return list_of_prime

def assign_dict_char_prime():
    dict_={}
    store_primes=prime_list()
    store_chars=[chr(x).lower() for x in range(65,91)]
    # dict_=dict(zip(store_chars,store_primes))
    for i in range(len(store_primes)):
        dict_[store_chars[i]]=store_primes[i]
    return dict_
assign_dict_char_prime()

def anagram_check(string1,string2):
    check_from_dict=assign_dict_char_prime()
    p1,p2=1,1
    for i in string1:
        if i in check_from_dict:
            p1=p1*check_from_dict[i]
            # print(p1)
    for i in string2:
        if i in check_from_dict:
            p2=p2*check_from_dict[i]
            # print(p2)

    if p1==p2:
        print("Strings are anagram of each other")
    else:
        print("Strings are not anagram of each other")
anagram_check("geeks","keegs")

    
```

```python
#WAP to flatten a nested list using recursion
def flatten(list1):
    final_list=[]
    for i in list1:
        if type(i)==list:
            final_list+=flatten(i)
        else:
            final_list.append(i)
    return final_list

flatten([1,2,3,[4,5],[5]])
```

```python

def binary_search(list_of_values,target):

    new_list=sorted(list_of_values)
    left=0
    right=len(new_list)-1
    while left<=right:
        mid=(left+right)//2
        if target==new_list[mid]:
            return mid
        elif target>new_list[mid]:
            left=mid+1
        else:
            right=mid-1
    return -1

binary_search([1,2,3,4,5],2)


```
