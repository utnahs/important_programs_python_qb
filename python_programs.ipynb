```python
1# WAP to enter the word in which you want to look for repetition of specified character
def count_char(char,word):
    c=0
    for _ in word:
        if char==_:
            c+=1
    return c

word=input("Enter the word in which you want to look for repetition of specified character")
char=input("Enter the character you want to look for")
count=count_char(char,word)
print(f"The character {char} appears {count} times in {word}")

```

```python
#WAP to print the multiplication table of a given number
def multiplication_table(num):
    i=1
    while i<=10:
        val=num*i
        yield val
        i+=1
s=int(input("Enter the number \n"))
for i in multiplication_table(s):
    print(i)
    
```

```python
#WAP to print the reverse of a number
def rev(num):
    s=0
    while num!=0:
        r=num%10
        s=s*10+r
        num=int(num/10)
    return s

if rev(123)==123:
    print("Palindrome")
else:
    print("Not Palindrome")
    
```

```python
#WAP to check if a number is prime number
def prime_check(num):
    c=0
    for i in range(1,num+1):
        if num%i==0:
            c+=1
    if c==2:
        return True
    else:
        return False


if prime_check(91)==True:
    print("Prime Number")
else:
    print("Not Prime")
```

```python
#Using the above function to print all theprime numbers till 100
for i in range(1,20):
    if prime_check(i):
        print(i)
```

```python
# WAP to print fibonacci series till specified range
def fibonacci(n):
    a=0
    b=1
    print(a)
    print(b)
    for i in range(0,n+1):
        c=a+b
        yield c
        a=b
        b=c
for i in fibonacci(6):
    print(i)
```

```python
#WAP to check if a number is armstrong
def armstrong_check(num):
    s=0
    pow=len(str(num))
    while num!=0:
        r=num%10
        s=s+r**pow
        num=int(num/10)
    return s

if armstrong_check(121)==121:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")



```

```python
# WAP to find the maximum and minimum of elements in a list
def find_max_min(list_):
    max=list_[0]
    min=list_[0]
    for num in list_:
        if num>max:
            max=num
        elif num<min:
            min=num
    return max,min  # return both values as a tuple by default
    #return {"Maximum": max_val, "Minimum": min_val} to return as dictionary


max,min=find_max_min([1,5,9,3,10,1,11])
print(f"The maximum is {max}")
print(f"The minimum is {min}")
```

```python
# WAP to remove duplicates from a list
def rem_dup(old_list):
    new_list=[]
    for item in old_list:
        if item not in new_list:
            new_list.append(item)
    return new_list
        
rem_dup([1,2,3,3,1])
```

```python
# WAP to remove duplicates from a list (not using membership operator)
def remove_duplicates(old_list):
    new_list=[]
    for i in range(len(old_list)):
        is_duplicate=False
        for j in range(len(new_list)):
            if old_list[i]==new_list[j]:
                is_duplicate=True
        if is_duplicate==False:
            new_list.append(old_list[i])
    return new_list

remove_duplicates([1,1,2,3,2])       
```

```python
#WAP to sort a list in ascending order
def sort_list(list_):
    temp = 0
    for i in range(len(list_)-1):
        for j in range(len(list_)-1-i):
            if list_[j]>list_[j+1]:
                temp = list_[j]
                list_[j]=list_[j+1]
                list_[j+1]=temp
    return list_

sort_list([1,4,3])
```

```python
# WAP to input a string and reverse it
def str_rev(input):
    reversed_string=''
    length=strlen(input)-1
    while length>=0:
        reversed_string=reversed_string+input[length]
        length-=1
    return reversed_string

str_rev("hello")

```

```python
# WAP to find the average of all numbers in a list
def avg(input):
    len=0
    sum=0
    for i in input:
        len+=1
        sum+=i
    return round((sum/len),2)
avg([1,2,3,4])   
```

```python
#WAP to find the sample standard deviation of given numbers in list
def stddev(listt):
    count=0
    mean=avg(listt)
    sq_diff_sum = 0
    for i in listt:
        count+=1
        sq_diff_sum+=(i-mean)**2
    s=(sq_diff_sum/(count-1))**0.5
    return s
stddev([1,2,3,4,5])
```

```python
# WAP to find the mode of a list

def frequency(user_input):                                      #function to find the frquency of each item in list
    unique_item_list=remove_duplicates(user_input)
    dict={}
    for i in range(len(unique_item_list)):
        c=0
        for j in range(len(user_input)):
            if unique_item_list[i] == user_input[j]:
                c+=1                                                 # frequencies
        dict[str(unique_item_list[i])]=c                             # storing the items along with its frequency in key value pair
    return dict

dict_=frequency([1,1,1,2,2,2])
print(dict_)
temp=0
for key in dict_:
    if dict_[key] > temp:
        temp=dict_[key]                                             # the max frequency
modes=[]
for key,value in dict_.items():                    #key and value unpacking  using items(), returns a view object containing keyâ€“value pairs of the dictionary.
    if value==temp:                                     # using the max frequency as lookup for the keys matching the maximum frequency as multiple modes could be present
        modes.append(key)
print(f"The mode/modes are {modes}")                
```
